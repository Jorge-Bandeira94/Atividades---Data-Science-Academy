```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path)
df
```

## Número de veículos pertencentes a cada marca

```python
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
# seaborn.factorplot() método é usado para desenhar um gráfico categórico em um FacetGrid.
# kind: (opcional) este parâmetro leva um valor de string, o tipo de gráfico a ser desenhado (corresponde ao nome de uma função de plotagem categórica. As opções são: “ponto”, “barra”, “tira”, “enxame”, “caixa” , “Violino” ou “boxen”.


g = sns.factorplot(y="brand", data=df, kind="count", palette="Reds_r", size=7, aspect=1.5)
g.ax.set_title("Veículos Por Marca",fontdict={'size':18})
g.ax.xaxis.set_label_text("Número de Veículos",fontdict= {'size':16})
g.ax.yaxis.set_label_text("Marca",fontdict= {'size':16})
plt.show()
```

```python
# Salvando o plot
g.savefig(("plots/Analise2/brand-vehicleCount.png"))
```
